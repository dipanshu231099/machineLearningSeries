```python
# Import libraries and classes required for this example:
import pandas as pd
from sklearn.datasets import load_iris

dataset = load_iris()
raw_data = dataset.data
names = dataset.feature_names
label = dataset.target 

# Convert dataset to a pandas dataframe:
data = pd.DataFrame(raw_data, columns=names)
# Use head() function to return the first 5 rows: 
data.head()
```

```python
label, dataset.target_names
```

```python
import matplotlib.pyplot as plt
colors = ["navy", "turquoise", "darkorange"]
for i in range(len(data)):
#     print(data.iloc[i][0],data.iloc[i][1],colors[label[i]])
    plt.scatter(data.iloc[i][0],data.iloc[i][1],color=colors[label[i]])
plt.show()
```
